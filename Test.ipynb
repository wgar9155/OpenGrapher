```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sin, lambdify

# Define the sympy function
x = symbols('x')
f = sin(x)/x

# Convert the sympy function to a numpy function
f_np = lambdify(x, f, modules='numpy')

# Generate x values
x_vals = np.linspace(-10, 10, 100)

# Evaluate the function at x values
y_vals = f_np(x_vals)

# Plot the function
plt.plot(x_vals, y_vals)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the sympy function')
plt.show()
```
